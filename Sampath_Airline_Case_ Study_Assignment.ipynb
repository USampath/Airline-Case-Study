# Reading Operations Sheet From Excel

```python
#Reading Data frame using Pd.read_excel
#using sheet_name parameter for reading a particular sheet in excel
#using header to remove unwanted empty rows and comment rows
#using header instead of skiprows because skip rows doesnt change the shape of data frame
Op = pd.read_excel("Data Science Case study Vindiata.xls",sheet_name = "Operations",header = 3)

#Dropping Unnecesary columns
Op.drop(Op.columns[0], axis = 1, inplace = True)

Op
```

```python
#To view the sahpe of datafarame i.e (no.of rows , no.of columns)
Op.shape
```

```python
# Creating new column in Op 'Total Hours flown' i.e sum of hours of all months in a row
Op['Total hours flown'] = Op.iloc[:,2:].sum(axis=1)
Op
```

```python
#Creating a copy of Op i.e Operation for further use
Operations = Op.copy()

#Calculating total hours flown by each aircraft using groupby
Operations = Operations.groupby(['Aircraft Type']).sum()
Operations
```

```python
#Joining two data frames using Merge to join both operations and Ac using common column=Aircraft Type
# we can use left or right or inner because we have same values for aircraft type
Ac_Op = Ac.merge(Operations , on = 'Aircraft Type' , how = 'left')
Ac_Op
```

```python
Ac_Op[['Aircraft Type','Range (Km)','Number of Seats','Costs per flight hour','Total hours flown']]
```

```python
# Findinng the total cost by aircraft type for the year
# Total cost = Total hours flown * Costs per flight hour

Ac_Op['Total Cost'] = Ac_Op['Total hours flown'] * Ac_Op['Costs per flight hour']
Ac_Op[['Aircraft Type','Total hours flown','Total Cost']]
```

```python
#Viewing required columns and sorted by total cost
Ac_Op[['Aircraft Type','Range (Km)','Number of Seats','Costs per flight hour','Total hours flown','Total Cost']].sort_values(['Total Cost'],ascending=False)
```

# We can see the Total Cost of every Aircraft type for the year

- Here A330 has flown more hours and it has the higest cost
- B747 has more capacity in seating and Range it stands second in cost 
- ATR72 stands last in Cost but it has very special features like seating capacity is 75 and Range is 1000 Km we can use on those conditions satisfy it is most useful for short trips with less people

# Which aircraft type  has the lowest cost per seat per km flown?

- Cost per seat Per Km = Total Cost / (no.of seats * total hours flown * Average Speed )

```python
# Creating new column Cost per seat per Km using above formula
Ac_Op['Cost per seat per Km'] = Ac_Op['Total Cost'] / (Ac_Op['Number of Seats']*Ac_Op['Ave. Speed (km/h)']*Ac_Op['Total hours flown'])

Ac_Op[['Aircraft Type','Cost per seat per Km']].sort_values(['Cost per seat per Km'])
```

# Here we can See that A330 has the lowest cost per seat per Km
- it is the most econmical Aircraft Among all 


```python
#Reading Data frame using Pd.read_excel
#using sheet_name parameter for reading a particular sheet in excel
#using header to remove unwanted empty rows and comment rows
#using header instead of skiprows because skip rows doesnt change the shape of data frame
Cp = pd.read_excel("Data Science Case study Vindiata.xls",sheet_name = "City pairs",header = 3)

#Dropping Unnecesary columns
Cp.drop(Cp.columns[0], axis = 1, inplace = True)
Cp
```

```python
#Creating a Copy of Ac_Op as df for futher evaluation
df = Ac_Op.copy()

#Dropping columns that doesnt require for Evaluation
df.drop(df.iloc[:, 5:19], axis = 1, inplace = True)
df
```

```python
#Dropping columns that doesnt require for Evaluation
df.drop(df.columns[[2,4]] , axis = 1 , inplace = True)
df
```

# Finding the Best Aircrafts for Different city Pairs
- First we check whether aircraft range is greater than the Distance between Citys
- Then we calculate Cost for trip



```python
# total cost of trip = cost per seat per Km * no.of Seats * no.of trips * Trip distance
# no.of trips = pass demand per day/no.of seats
# Creating new columns in df using for loop using itertuple
# if condition doesnt satisfy it will give "NaN"

for a,b,c,d in Cp.itertuples(index=False):
    df[str(a)+'-'+str(b)+' cost'] = np.where(df['Range (Km)'] >= d, (df['Cost per seat per Km'] * df['Number of Seats'] * d * np.ceil(c / df['Number of Seats'])).astype(int), "NaN")

# Display the dataframe with added travel cost
df
```
